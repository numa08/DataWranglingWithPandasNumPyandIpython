```python
tips = pd.read_csv('../examples/tips.csv')
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
```

```python
party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]
```

```python
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
```

```python
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
macro = pd.read_csv('../examples/macrodata.csv')
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha', 0.2})
```

```python
sns.factorplot(x='day', y='tips_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tips_pct < 1])
```

```python
sns.factorplot(x='day', y='tips_pct', row='time', col='smoker', kind='bar', data=tips[tips.tips_pct < 1])
```

```python
sns.factorplot(x='tips_pct', y='day', kind='box', data=tips[tips.tips_pct < 0.5])
```
